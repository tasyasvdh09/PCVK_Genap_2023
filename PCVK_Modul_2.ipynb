**create black image bentuk geometri 2D dari OpenCV, tipe data int16**

```python
black_img = np.zeros(shape=(512,512,3),dtype=np.int16)
plt.imshow(black_img)
```

**create black image dgn tipe data int32**

```python
black_img2=np.zeros(shape=(512,512,3),dtype=np.int32)
plt.imshow(black_img2)
```

**inisialisasi NumPy array dengan tipe data int32**

```python
vertices = np.array([[100,300],[200,200],[400,300],[200,400]],dtype=np.int32)
vertices
```

**reshape**

```python
pts = vertices.reshape((-1,1,2))  # nilai 2 untuk menunjukkan bahwa tiap titik dibuat 3 channel yg mewakili R,G,B
pts
```

Pertanyaan tambahan BGR RGB menggunakan OpenCV dan matplotlib R,G,B

```python
# Load the image
img = cv.imread('/content/drive/MyDrive/kitten.jpeg') # Import gambar

# Split the image into its R, G, and B channels
b, g, r = cv.split(img)

# Display the R, G, and B channels as grayscale images
cv2_imshow(r)
cv2_imshow(g)
cv2_imshow(b)
```

```python
# Create a blank image of zeros
zeros = np.zeros(img.shape[:2], dtype = "uint8")

# Merge the Red channels back together
red_image = cv.merge([zeros, zeros, r])

# Display the Red channels as grayscale image
cv2_imshow(red_image)
plt.imshow(red_image) # matplotlib menghasilkan blue karena matplotlib menghasilkan RGB, bukan BGR
```

```python
# Merge the Green channel back together
green_image = cv.merge([zeros, g, zeros])

# Display the Green channel as grayscale image
cv2_imshow(green_image)
plt.imshow(green_image)
```

```python
# Merge the Blue channel back together
blue_image = cv.merge([b, zeros, zeros])

# Display the Blue as grayscale image
cv2_imshow(blue_image)
plt.imshow(blue_image) # matplotlib menghasilkan blue karena matplotlib menghasilkan RGB, bukan BGR
```
