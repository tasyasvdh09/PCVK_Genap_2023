```python
# import library
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

**1.) Lakukan Face Detection untuk image object lain yang tersedia pada (/images/facedet).**

```python
cascade_wajah = cv.CascadeClassifier('/content/drive/MyDrive/PCVK/fitur-facedet/haarcascade_frontalface_alt.xml')

jokowi = cv.imread('/content/drive/MyDrive/PCVK/images/jokowi.jpg',0)
jokowi_rgb = cv.cvtColor(jokowi, cv.COLOR_BGR2RGB)
jokowi_fd = jokowi_rgb.copy()

fd_jokowi = cascade_wajah.detectMultiScale(jokowi_fd)
for(x,y,w,h) in fd_jokowi:
  cv.rectangle(jokowi_fd, (x,y),(x+w,y+h), (0,255,0), 5)
plt.figure(figsize = (10,10))
f, axarr = plt. subplots(1,2,figsize=(8,8))
plt.subplot(121),plt.imshow(jokowi_rgb,cmap='gray')
plt.subplot(122),plt.imshow(jokowi_fd,cmap='gray')
plt.show()
```

**2.) Wajah kucing tidak bisa dideteksi dengan baik. Lakukan deteksi wajah kucing hingga muncul rectangle pada bagian wajahnya seperti gambar berikut. Petunjuk pada soal ini, perhatikan pretrained features yang telah disediakan OpenCV. Gunakan xml yang ada jika memang telah disediakan.**

```python
cascade_wajah = cv2.CascadeClassifier('/content/drive/MyDrive/PCVK/fitur-facedet/haarcascade_frontalcatface.xml')

cat = cv2.imread('/content/drive/MyDrive/PCVK/images/mycat.jpg')

# Clone the original image for displaying the detected faces
detected_faces = cat.copy()

roi_wajah = cascade_wajah.detectMultiScale(cat)
for (x, y, w, h) in roi_wajah:
    cv2.rectangle(detected_faces, (x, y), (x+w, y+h), (0, 255, 0), 3)

# Create subplots to display original and detected faces
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot the original image in grayscale
axes[0].imshow(cv2.cvtColor(cat, cv2.COLOR_BGR2GRAY), cmap='gray')
axes[0].set_title('Original Image')

# Plot the image with detected faces in grayscale
axes[1].imshow(cv2.cvtColor(detected_faces, cv2.COLOR_BGR2GRAY), cmap='gray')
axes[1].set_title('Detected Faces')

# Hide the axis labels
for ax in axes:
    ax.axis('off')

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```
