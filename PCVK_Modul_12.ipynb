```python
# import library
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

**1.) Lakukan Face Detection untuk image object lain yang tersedia pada (/images/facedet).**

```python
cascade_wajah = cv2.CascadeClassifier('/content/drive/MyDrive/PCVK/fitur-facedet/haarcascade_frontalface_alt.xml')

jokowi = cv2.imread('/content/drive/MyDrive/PCVK/images/jokowi.jpg')

# Clone the original image for displaying the detected faces
detected_faces = jokowi.copy()

roi_wajah = cascade_wajah.detectMultiScale(jokowi)
for (x, y, w, h) in roi_wajah:
    cv2.rectangle(detected_faces, (x, y), (x+w, y+h), (0, 255, 0), 3)

# Create subplots to display original and detected faces
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot the original image
axes[0].imshow(cv2.cvtColor(jokowi, cv2.COLOR_BGR2RGB))
axes[0].set_title('Original Image')

# Plot the image with detected faces
axes[1].imshow(cv2.cvtColor(detected_faces, cv2.COLOR_BGR2RGB))
axes[1].set_title('Detected Faces')

# Hide the axis labels
for ax in axes:
    ax.axis('off')

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```
